```python
def is_palindrome(n): 
    original = n 
    reverse = 0 
    while n > 0: 
        reverse = reverse * 10 + n % 10 
        n //= 10 
        return original == reverse 
num = int(input("Enter a number: ")) # Taking input from user 
print(is_palindrome(num))
```
