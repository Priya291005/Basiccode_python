```python
a = input()

for i in range(len(a)):
    count = 0
    for j in range(len(a)):
        if a[i] == a[j]:
            count += 1
    if count > 1 and a[i] not in a[:i] and a[i] != " ":
        print(a[i])
```

```python
a = input().lower()   # case problem avoid

d = {}

for i in a:
    if i != " ":      # space ignore
        if i in d:
            d[i] += 1
        else:
            d[i] = 1

printed = []

for i in a:           # order maintain
    if i in d and d[i] > 1 and i not in printed:
        print(i)
        printed.append(i)
```
