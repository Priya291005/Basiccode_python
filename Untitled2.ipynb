```python
s=input()
r=""
for i in range (len(s)-1,-1,-1):
    r=r+s[i]
if(r==s):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
li=list(map(int,input("Enter values:").split()))
max=li[0]
smax=li[0]
for i in range(len(li)):
    if li[i]>max:
        smax=max
        max=li[i]
    elif li[i]>smax and li[i]<max:
        smax=li[i]
if smax==0:
    print("No second largest")
else:
    print(smax)
```

```python
n=int(input())
a=list(map(int, input().split()))
largest=max(a)
new_list=[]
for i in a:
    if i!=largest:
        new_list.append(i)
if len(new_list) == 0:
    print("No second largest")
else:
    print(max(new_list))
```
