```python
"""
1.Bus Ticket Booking
A bus company records the number of passengers boarding at different stops during a trip.
Write a program to calculate the total number of passengers in the bus after the last stop.
Rules
At each stop:
	Some passengers get down
	Some passengers get in
The bus starts with 0 passengers.
Input
First line → Number of stops N
Next N lines contain two integers:
out in
•	out → passengers getting down
•	in → passengers getting in
INPUT :
4
0   10
3    5
4    2
2    3
OUTPUT :
11
Explanation
•	Stop 1 → 0 out, 10 in → total = 10
•	Stop 2 → 3 out, 5 in → total = 12
•	Stop 3 → 4 out, 2 in → total = 10
•	Stop 4 → 2 out, 3 in → total = 11
"""

n=int(input())
b=0
for i in range(n):
    o,i=map(int,input().split())
    b=(b-o)+i
print(b)
```

```python
"""
2.Corrupted Login Log
Scenario
A system log records login attempts as a string:
•	S → Successful login
•	F → Failed login
If 3 consecutive failed logins occur, the account must be locked.

Write a program to determine whether the account will be locked or not.
Input
LoginLog = "SFFFSFSF"
Explanation
Sequence:
S F F F S F S F
There are 3 consecutive F → Account locked.
Output
Account Locked

Example 2
Input
LoginLog = "SFFSFSF"

Output
Account Safe

"""

n=input()
if n.isupper():
    if 'FFF'in n:
        print("Locked")
    else:
        print("Safe")
else:
    print("Invalid")
```

```python
"""
3. Word Mirror Checker
Scenario
A string is called a Mirror Word if:
•	The first half of the string is the same as the reverse of the second half.
Input
String = "abccba"
Output
Mirror String






Example 2	
Input: "abcdef"
Output: Not Mirror

Explanation
First half → abc
Second half → cba
Reverse of second half → abc
Both match → Mirror word.

"""
n= input()
l = len(n) // 2
s = n[:l]
r = n[l:]  
print("MIRROR WORD") if s == r[::-1] else print("NOT MIRROR")
```

```python
"""
4. Classroom Height Visibility Problem
Scenario
In a classroom, N students are standing in a line from left to right.
Each student has a height represented in an array.
A student can see the board clearly only if no taller student is standing in front of them (towards the left).
Your task is to find how many students can see the board.
Input Specification
•	input1: An integer N representing number of students
•	input2: An integer array heights[] representing the height of each student
Output Specification
Return an integer representing the number of students who can see the board.
INPUT :
6
5 3 8 3 6 9
OUTPUT :
3
"""
n=int(input())
s=list(map(int,input().split()))
m=s[0]
c=1
for i in range(1,n):
    if s[i]>m:
        c+=1
        m=s[i]
print(c)
```

```python
"""
A sports tournament has N players. Each player has a cube jersey number (1³, 2³, 3³ ... N³). Before the match starts, the coach checks the jersey list. Suddenly he notices one player's jersey number is MISSING from the list!
Example
Input :
5
1 8 64 125 0
Output :
27
"""
n=int(input())
l=list(map(int,input().split()))
c=[]
for i in range(1,n+1):
    p=i**3
    if l[i-1]!=p:
        c.append(p)
        break
print(c)
```

```python
"""Leaders in Array 
A number is a leader if no number on its right side is greater than it.
Simple:
If a number is greater than all numbers after it, it is a leader.
Example
Input:
6
[16, 17, 4, 3, 5, 2]
We check each number and see right side.
OUTPUT:

17 5 2

Explanation :
Step-by-Step
Check 16
Right side = [17, 4, 3, 5, 2]
17 is bigger than 16   =Not leader

Check 17
Right side = [4, 3, 5, 2]
No number is bigger than 17 = Leader

Check 4
Right side = [3, 5, 2]
5 is bigger than 4 = Not leader

Check 3
Right side = [5, 2]
5 is bigger =Not leader

Check 5
Right side = [2]
No number bigger =Leader

Check 2
Right side = []
No elements  =Leader (Last element always leader)

Final Leaders
Output:
17 5 2
"""
n=int(input())
l=list(map(int,input().split()))
for i in range(len(l)):
    m=1
    for j in range(i+1,len(l)):
        if l[i]<l[j]:
            m=0
            break
    if m==1:
        print(l[i],end=" ")
```

```python
"""
Expected Input
You will be given:
1.	Number of patients n 
2.	Entry time array 
3.	Exit time array 
Example Input
Enter number of patients:
4

Enter entry times:
1 3 5 7

Enter exit times:
2 6 8 10

Expected Output
Maximum beds required = 2

Another Example
Input
Enter number of patients:
5

Entry:
1 2 3 4 5

Exit:
2 3 4 5 6
Output
Maximum beds required = 1
(Only one patient at a time)

Team Selection
Question:
Select 3 numbers such that difference between max and min is minimum.
Input:
7
2 3 4 7 9 12 56
3

Output:
2

EXPLANATION :
Packets	            Max	     Min	        Difference
2,3,4	             4	       2	             2
3,4,7	             7	       3	             4
4,7,9	             9	       4	             5
7,9,12	            12	       7	             5
9,12,56	            56	       9	            47

So minimum difference is 2
"""
n=int(input())
l=list(map(int,input().split()))
s=int(input())
c=[]
for i in range(n-s+1):
    a=l[i:s+i]
    minn=min(a)
    maxx=max(a)
    d=maxx-minn
    c.append(d)
print(min(c))
```
