```python
import mysql.connector
db=mysql.connector.connect(
    host="127.0.0.1",
    username="root",
    password="Priya@2910",
    database="rock"
)
cursor=db.cursor()
cursor.execute("""create table st(Id int,Name varchar(30),City varchar(20))""")
while (True):
    print("1.Insert Student data")
    print("2.View Student data")
    print("3.Exit")
    print("Enter your choice:")
    n=int(input())
    if n==1:
        print("Enter the count of insert data:")
        n=int(input())
        for i in range(1,n+1):
            cursor=db.cursor()
            Id=int(input("Enter id:"))
            Name=input("Enter name:")
            City=input("Enter city:")
            sql="insert into st values(%s,%s,%s)"
            cursor.execute(sql,(Id,Name,City))
            db.commit()
            print("Data insert successfullly")
    elif n==2:
        cursor=db.cursor()
        cursor.execute("select * from st")
        r=cursor.fetchall()
        for i in r:
            print(i)
    elif n==3:
        break
    else:
        print("Invalid choice")
```
