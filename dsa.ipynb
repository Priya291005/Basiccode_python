```python
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif key < arr[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return -1

arr = [12, 23, 34, 45, 78, 90]
print(binary_search(arr, 78))
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):              # Number of passes
        for j in range(n - 1 - i):      # Comparisons in each pass
            if arr[j] > arr[j + 1]:     # Swap if elements are in wrong order
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

arr = [45, 12, 78, 34, 23, 90]
print("Sorted Array:", bubble_sort(arr))
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

arr = [45, 12, 78, 34, 23, 90]
print(insertion_sort(arr))
```

```python
def powers(b,e):
    if e==0:
        return 1
    return b*powers(b,e-1)
b=int(input())
e=int(input())
print(powers(b,e))
```

```python
def coun(n,c):
    if n==0:
        return c
    return coun(n//10,c+1)
n=int(input())
print(coun(n,0))
```
